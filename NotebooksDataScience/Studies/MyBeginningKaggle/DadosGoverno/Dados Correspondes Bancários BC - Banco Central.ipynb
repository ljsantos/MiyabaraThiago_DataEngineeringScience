```python
dfnew = (json_normalize(df.value))
dfnew.head()
```

```python
cUrlMunicipios = "https://servicodados.ibge.gov.br/api/v1/localidades/municipios?$format=json"
#"https://servicodados.ibge.gov.br/api/v1/localidades/estados/{UF}/municipios"
municipios = pd.read_json(cUrlMunicipios)
dfCities = pd.DataFrame(municipios) 
dfCities.head()
```

```python
dfCities.rename(columns={'id': 'MunicipioIBGE', 'nome': 'Municipio'}, inplace=True)
dfCities = dfCities.drop('microrregiao', axis=1)
dfCities.head()
```

```python
dfCities.info()
```

```python
dfCities["MunicipioIBGE"] = pd.to_numeric(dfCities["MunicipioIBGE"])
```

```python
dfnew.merge(dfCities, left_on='MunicipioIBGE', right_on='MunicipioIBGE')
```
